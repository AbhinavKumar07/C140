```python
import pandas as pd

credits_df = pd.read_csv('tmdb_5000_credits.csv')
movies_df = pd.read_csv('tmdb_5000_movies.csv')

credits_df.head(10)
```

```python
credits_df.info()
```

```python
print(credits_df.shape)
print(movies_df.shape)
```

```python
credits_df.rename(columns={"movie_id":"id"}, inplace=True ) #Doesnt duplicate dataframe
credits_df.head()
```

```python
results_df = pd.merge(movies_df , credits_df , on="id")

results_df.shape
```

```python
#C139

results_df.info()
```

```python
results_df.drop(["homepage" , 'title_x' , 'title_y' , 'production_companies' ] , axis = 1 , inplace = True)
```

```python
results_df.info()
```

```python
results_df.dropna(inplace = True)
results_df.info()
```

```python
R = results_df['vote_average']
V = results_df['vote_count']
C = results_df['vote_average'].mean()
print(C)
```

```python
M = results_df['vote_count'].quantile(0.9)
print(M)
```

```python
results_df['weighted_rating'] = (R*V + C*M) / (V + M)
results_df[['original_title' , 'weighted_rating']].head()
```

```python
results_df.sort_values('weighted_rating' , ascending = False , inplace = True)
results_df[['original_title' , 'vote_average' , 'vote_count' , 'weighted_rating', 'popularity']].head(10)
```

```python
import plotly.express as px

bar_plot = px.bar(results_df.head(10).sort_values('weighted_rating'),x = 'weighted_rating' , y = 'original_title' , orientation='h')
bar_plot.show()
```

```python
from ast import literal_eval

features = ['cast', 'crew' , 'genres' , 'keywords']

for feature in features:
  results_df[feature] = results_df[feature].apply(literal_eval)

type(results_df.loc[0].at['cast'])
```

```python
results_df.loc[6].at['crew']
```

```python
import numpy as np

def get_director(crew):
  for crew_member in crew:
    if crew_member['job'] == 'Director':
      return crew_member['name']
  return np.nan

results_df['director'] = results_df['crew'].apply(get_director)

results_df[['original_title','director',]].head()
```

```python
def get_name_list(column_value):
  names_list = []

  if isinstance(column_value , list):
    for element in column_value:
      names_list.append(element['name'])
  return names_list

features = ['cast' , 'keywords' , 'genres']

for feature in features:
  results_df[feature]  = results_df[feature].apply(get_name_list)

results_df[['cast' , 'keywords' , 'genres' , 'director']].head()
```
